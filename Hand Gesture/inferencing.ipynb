```python
def preprocess_image(frame):
    
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    resized_frame = cv2.resize(gray_frame, (320, 120))
    normalized_frame = resized_frame / 255.0
    processed_frame = normalized_frame.reshape(1, 120, 320, 1)
    
    return processed_frame

def predict_gesture(frame):
    preprocessed_frame = preprocess_image(frame)
    prediction = model.predict(preprocessed_frame)
    predicted_class = np.argmax(prediction)
    return predicted_class
```

```python
lookup = dict()
reverselookup = dict()
count = 0
for j in os.listdir('D:\\mp\\00'):
    if not j.startswith('.'): # If running this code locally, this is to 
                              # ensure you aren't reading in hidden folders
        lookup[j] = count
        reverselookup[count] = j
        count = count + 1
print(lookup)
```

```python
import cv2
import numpy as np
from keras.models import load_model

# Load your trained model
model = load_model('recognizer.h5')

# Define the gesture labels
gesture_labels = [reverselookup[i] for i in range(10)]

# Initialize your laptop camera
camera = cv2.VideoCapture(0)  # 0 for the default camera, change if needed

while True:
    ret, frame = camera.read()
    if not ret:
        break

    # Preprocess the image
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    resized_frame = cv2.resize(gray_frame, (320, 120))
    normalized_frame = resized_frame / 255.0
    processed_frame = normalized_frame.reshape(1, 120, 320, 1)

    # Make a prediction
    prediction = model.predict(processed_frame)
    predicted_class = np.argmax(prediction)
    gesture_label = gesture_labels[predicted_class]

    # Display the recognized gesture label on the video feed
    cv2.putText(frame, gesture_label, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

    cv2.imshow("Gesture Recognition", frame)

    # Press 'q' to exit the video feed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

camera.release()
cv2.destroyAllWindows()
```
